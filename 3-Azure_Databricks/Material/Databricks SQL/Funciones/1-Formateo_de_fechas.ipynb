````python
USE f1_processed;
````

````python
DESCRIBE FORMATTED drivers;
````

#### Formato 1

````python
SELECT *, DATE_FORMAT(dob, 'dd-MM-yyyy')
  FROM drivers
````

````python
CREATE TABLE f1_processed.drivers_date_format1
AS
SELECT *, DATE_FORMAT(dob, 'dd-MM-yyyy') date_format
  FROM f1_processed.drivers
````

````python
SELECT * FROM f1_processed.drivers_date_format1
````

````python
DESCRIBE FORMATTED f1_processed.drivers_date_format1;
````

#### Formato 2

````python
CREATE TABLE f1_processed.drivers_date_format2
AS
SELECT *, FROM_UNIXTIME(UNIX_TIMESTAMP(dob,'yyyy-MM-dd'),'dd-MM-yyyy') date_format
  FROM f1_processed.drivers
````

````python
SELECT * FROM f1_processed.drivers_date_format2
````

````python
DESCRIBE FORMATTED f1_processed.drivers_date_format2;
````

#### Formato 3

````python
CREATE TABLE f1_processed.drivers_date_format3
AS
SELECT *, TO_DATE(FROM_UNIXTIME(UNIX_TIMESTAMP(date_format , 'dd-MM-yyyy'))) date
  FROM f1_processed.drivers_date_format2
````

````python
%sql
SELECT * FROM f1_processed.drivers_date_format3
````

````python
DESCRIBE FORMATTED f1_processed.drivers_date_format3;
````

#### Formato 4

````python
CREATE TABLE f1_processed.drivers_date_format4
AS
SELECT *, REPLACE(dob, '-', '/') date_format
  FROM f1_processed.drivers
````

````python
SELECT * FROM f1_processed.drivers_date_format4
````

````python
DESCRIBE FORMATTED f1_processed.drivers_date_format4;
````
