```python
USE demo;
SHOW TABLES;
```

```python
DESCRIBE EXTENDED race_results_python;
```

```python
DESCRIBE FORMATTED race_results_python;
```

```python
SELECT *
  FROM demo.race_results_python
 WHERE race_year = 2020;
```

```python
SELECT CURRENT_DATABASE()
```

```python
CREATE TABLE demo.race_results_sql
AS
SELECT *
  FROM demo.race_results_python
 WHERE race_year = 2020;
```
