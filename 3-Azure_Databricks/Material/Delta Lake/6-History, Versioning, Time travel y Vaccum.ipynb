#### **Vaccum**

Nos permite eliminar la habilidad para viajar en el tiempo para las versiones

```python
VACUUM f1_demo.drivers_merge
```

```python
-- Corresponde a la versión 3
SELECT * FROM f1_demo.drivers_merge TIMESTAMP AS OF '2023-06-16T23:56:48.000+0000';
```

```python
-- Vaccum por defecto elimina la habilidad para viajar en el tiempo para las versiones que lleven 168 horas (7 dias) desde su ejecución
-- Si queremos hacerlo para una versión realizada antes de los 7 dias, podemos realizar lo siguiente:
SET spark.databricks.delta.retentionDurationCheck.enabled = false;
-- Al utilizar RETAIN 0 HOURS se eliminara la habilidad para viajar en el tiempo para todas las versiones de manera inmediata
-- Por ejemplo, si colocasemos RETAIN 20 HOURS, mantendremos la habilidad para viajar en el tiempo para las versiones ya ejecutadas que lleven menos 20 horas desde su ejecución
VACUUM f1_demo.drivers_merge RETAIN 0 HOURS  
```

```python
-- Vemos que se nos devuelve un error al tratar de viajar en el tiempo para una versión especifica
SELECT * FROM f1_demo.drivers_merge TIMESTAMP AS OF '2023-06-16T23:36:10.000+0000';
```

```python
-- La data del dataframe sigue intacta, el utilizar VACCUM no la afecta en lo absoluto
SELECT * FROM f1_demo.drivers_merge
ORDER BY driverId;
```

```python
-- La Historia tampoco se ve afectada por el uso de VACCUM
-- VACCUM nos permite configurar el inhabilitar poder viajar en el tiempo
DESC HISTORY f1_demo.drivers_merge;
```

##### Recuperar un registro utilizando versiones

```python
SELECT * FROM f1_demo.drivers_merge
ORDER BY driverId;
```

```python
DELETE FROM f1_demo.drivers_merge WHERE driverId = 1;
```

```python
SELECT * FROM f1_demo.drivers_merge
ORDER BY driverId;
```

```python
-- En la version 3 vemos que teniamos todos los registros
SELECT * FROM f1_demo.drivers_merge VERSION AS OF 3
ORDER BY driverId;
```

```python
-- Realizamos un INSERT usando MERGE y apoyandonos en el dataframe correspondiente a la version 3 como tabla origen
MERGE INTO f1_demo.drivers_merge t
USING f1_demo.drivers_merge VERSION AS OF 3 s
   ON (t.driverId = s.driverId)
WHEN NOT MATCHED THEN
   INSERT *
```

```python
SELECT * FROM f1_demo.drivers_merge
ORDER BY driverId;
```

```python
-- De han agregado 2 nuevas versiones (4 y 5) correspondientes a las dos últimas acciones que realizamos: DELETE y MERGE
DESC HISTORY f1_demo.drivers_merge
```
