### ![Spark Logo Tiny](https://files.training.databricks.com/images/105/logo_spark_tiny.png) **Convertir un directorio Parquet en un directorio Delta**

Utilizaremos la tabla **f1_demo.drivers_convert_to_delta** que creamos en el punto **7-Convertir Parquet tables en Delta tables**

```python
df = spark.table("f1_demo.drivers_convert_to_delta")
```

Aqui estamos utilizando **save** lo cual nos crea un directorio y archivos. Al utilizar **saveAsTable** creamos una tabla y directorio y archivos

```python
df.write.format("parquet").save("/mnt/formula1dl/demo/drivers_convert_to_delta_new")
```
