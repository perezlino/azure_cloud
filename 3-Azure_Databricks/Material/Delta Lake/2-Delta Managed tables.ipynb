#### **Crear Database**

```python
CREATE DATABASE IF NOT EXISTS f1_demo
LOCATION '/mnt/formula1dl/demo'
```

#### **Leer archivo en un dataframe**

```python
results_df = spark.read \
.option("inferSchema", True) \
.json("/FileStore/2021-03-28/results.json")
#.json("/mnt/formula1dl/raw/2021-03-28/results.json")
```

#### **Crear DELTA MANAGED TABLES**

- Recordar que para las **MANAGED TABLES** con solo eliminar la tabla, se eliminará la tabla y el directorio y archivo creado donde fue almacenada

##### **Forma 1**

```python
# No le indicamos el formato --> .format("parquet")
# Se guardó como una tabla DELTA
results_df.write.mode("overwrite").saveAsTable("f1_demo.results_managed")
```

##### Forma 5

Si creamos una tabla utilizando **Databricks SQL** y no le indicamos un **LOCATION** o un **OPTIONS(path)** la tabla se creará como **MANAGED TABLE**

```python
results_df.dtypes
```

```python
CREATE TABLE IF NOT EXISTS f1_demo.results_merge(
constructorId INT,
driverId INT,
fastestLap STRING,
fastestLapSpeed STRING,
fastestLapTime STRING,
grid INT,
laps INT,
milliseconds STRING,
number INT,
points INT,
position STRING,
positionOrder INT,
positionText STRING, 
raceId INT, 
rank INT,
resultId INT,
statusId INT,
time STRING
)
USING DELTA -- No es necesario indicar que es DELTA, para que nuestra tabla sea DELTA
```

```python
results_df.createOrReplaceTempView('v_results_df')
```

```python
SELECT * FROM v_results_df;
```

```python
-- t: target y s: source
MERGE INTO f1_demo.results_merge t
USING v_results_df s
ON t.driverId = s.driverId
WHEN MATCHED THEN
  UPDATE SET t.constructorId = s.constructorId,
             t.driverId = s.driverId,
             t.fastestLap = s.fastestLap,
             t.fastestLapSpeed = s.fastestLapSpeed,
             t.fastestLapTime = s.fastestLapTime,
             t.grid = s.grid,
             t.laps = s.laps,
             t.milliseconds = s.milliseconds,
             t.number = s.number,
             t.points = s.points,
             t.position = s.position,
             t.positionOrder = s.positionOrder,
             t.positionText = s.positionText,
             t.raceId = s.raceId,
             t.rank = s.rank,
             t.resultId = s.resultId,
             t.statusId = s.statusId,
             t.time = s.time
WHEN NOT MATCHED
  THEN INSERT (t.constructorId,t.driverId,t.fastestLap,t.fastestLapSpeed,t.fastestLapTime,t.grid,t.laps,t.milliseconds,
               t.number,t.points,t.position,t.positionOrder,t.positionText,t.raceId,t.rank,t.resultId,t.statusId,t.time)
               
  VALUES (s.constructorId,s.driverId,s.fastestLap,s.fastestLapSpeed,s.fastestLapTime,s.grid,s.laps,s.milliseconds,
          s.number,s.points,s.position,s.positionOrder,s.positionText,s.raceId,s.rank,s.resultId,s.statusId,s.time)
```

```python
SELECT * FROM f1_demo.results_merge;
```

```python
DESCRIBE FORMATTED f1_demo.results_merge;
```
