### ![Spark Logo Tiny](https://files.training.databricks.com/images/105/logo_spark_tiny.png) **Delta Managed tables**

#### **Crear Database**

```python
CREATE DATABASE IF NOT EXISTS f1_demo
LOCATION '/mnt/formula1dl/demo'
```

#### **Leer archivo en un dataframe**

```python
results_df = spark.read \
.option("inferSchema", True) \
.json("/mnt/formula1dl/raw/2021-03-28/results.json")
```

#### **Crear DELTA MANAGED TABLES**

- Recordar que para las **MANAGED TABLES** con solo eliminar la tabla, se eliminará la tabla y el directorio y archivo creado donde fue almacenada

##### **Forma 1**

```python
# No le indicamos el formato --> .format("parquet")
# Se guardó como una tabla DELTA
results_df.write.mode("overwrite").saveAsTable("f1_demo.results_managed")
```

<center><img src="https://i.postimg.cc/bwv9h8hC/db164.png"></center>

```python
%sql
DESCRIBE FORMATTED f1_demo.results_managed;
```

```python
%sql
SELECT * FROM f1_demo.results_managed;
```

```python
%sql
DROP TABLE f1_demo.results_managed
```

##### **Forma 2**

```python
# Otra forma de guardar la tabla como DELTA
results_df.write.format("delta").mode("overwrite").saveAsTable("f1_demo.results_managed")
```

<center><img src="https://i.postimg.cc/hjYH9M0g/db165.png"></center>

```python
%sql
DESCRIBE FORMATTED f1_demo.results_managed;
```

```python
%sql
DROP TABLE f1_demo.results_managed
```

##### **Forma 3**

```python
# Al crear una tabla utilizando ".format("parquet")" no creamos una tabla DELTA
results_df.write.format("parquet").mode("overwrite").saveAsTable("f1_demo.results_managed")
```

<center><img src="https://i.postimg.cc/wT1rdSdk/db166.png"></center>

```python
%sql
DESCRIBE FORMATTED f1_demo.results_managed;
```

```python
%sql
DROP TABLE f1_demo.results_managed
```

##### Forma 4

```python
# Al crear una tabla utilizando ".partitionBy()" también creamos una tabla DELTA
results_df.write.partitionBy("constructorId").mode("overwrite").saveAsTable("f1_demo.results_managed")
```

```python
%sql
DESCRIBE FORMATTED f1_demo.results_managed;
```

```python
%sql
SHOW PARTITIONS f1_demo.results_managed
```

```python
%sql
DROP TABLE f1_demo.results_managed
```
