### ![Spark Logo Tiny](https://files.training.databricks.com/images/105/logo_spark_tiny.png) **Updates y Deletes en Delta tables**

#### **Crear Database**

```python
CREATE DATABASE IF NOT EXISTS f1_demo
LOCATION '/mnt/formula1dl/demo'
```

#### **Leer archivo en un dataframe**

```python
results_df = spark.read \
.option("inferSchema", True) \
.json("/mnt/formula1dl/raw/2021-03-28/results.json")
```

```python
# Se guard√≥ como una DELTA MANAGED TABLE
results_df.write.mode("overwrite").saveAsTable("f1_demo.results_managed")
```

```python
SELECT * FROM f1_demo.results_managed;
```

#### **Update en Delta tables**

##### Utilizando Databricks SQL

```python
UPDATE f1_demo.results_managed
  SET points = 11 - position
WHERE position <= 10
```

```python
SELECT * FROM f1_demo.results_managed;
```

##### Utilizando Python

Forma 1

```python
from delta.tables import DeltaTable

deltaTable = DeltaTable.forPath(spark, "/mnt/formula1dl/demo/results_managed")

deltaTable.update("position <= 10", { "points": "21 - position" } ) 
```

```python
SELECT * FROM f1_demo.results_managed;
```

Forma 2

```python
from delta.tables import DeltaTable

deltaTable = DeltaTable.forPath(spark, "/mnt/formula1dl/demo/results_managed")

deltaTable.update(condition = "position <= 10", 
                        set = { "points": "41 - position" } ) 
```

```python
SELECT * FROM f1_demo.results_managed;
```

#### **Delete en Delta tables**

##### Utilizando Databricks SQL

```python
DELETE FROM f1_demo.results_managed
WHERE position > 10;
```

```python
SELECT * FROM f1_demo.results_managed;
```

##### Utilizando Python

```python
from delta.tables import DeltaTable

deltaTable = DeltaTable.forPath(spark, "/mnt/formula1dl/demo/results_managed")

deltaTable.delete("points = 0") 
```

```python
SELECT * FROM f1_demo.results_managed;
```
