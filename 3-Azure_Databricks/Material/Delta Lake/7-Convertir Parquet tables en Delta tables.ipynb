Utilizaremos la tabla **f1_demo.drivers_merge** que creamos en el punto **5-Upsert usando MERGE**

#### Convertir Parquet tables en Delta tables

```python
CREATE TABLE IF NOT EXISTS f1_demo.drivers_convert_to_delta (
driverId INT,
dob DATE,
forename STRING, 
surname STRING,
createdDate DATE, 
updatedDate DATE
)
USING PARQUET
```

```python
INSERT INTO f1_demo.drivers_convert_to_delta
SELECT * FROM f1_demo.drivers_merge
```

```python
CONVERT TO DELTA f1_demo.drivers_convert_to_delta
```
