### **Notebook Basics**

Los notebooks son el medio principal para desarrollar y ejecutar código de forma interactiva en Databricks. Esta lección proporciona una introducción básica al trabajo con notebooks Databricks.

Si has usado anteriormente notebooks Databricks pero esta es la primera vez que ejecutas un notebook en Databricks Repos, notarás que la funcionalidad básica es la misma. En la siguiente lección, revisaremos algunas de las funcionalidades que Databricks Repos añade a los notebooks.

### Objetivos de aprendizaje
Al final de esta lección, deberías ser capaz de:

A lo largo de este curso, nuestros notebooks están diseñados como instrumentos de aprendizaje. Los notebooks pueden ser fácilmente desplegados como código de producción con Databricks, así como proporcionar un robusto conjunto de herramientas para la exploración de datos, informes y dashboards.

### **Comandos mágicos**
* Los comandos mágicos son específicos de los notebooks Databricks
* Son muy similares a los comandos mágicos que se encuentran en otros notebooks comparables
* Se trata de comandos integrados que proporcionan el mismo resultado independientemente del idioma del notebook
* Un símbolo de porcentaje (%) al principio de una celda identifica un comando mágico.
  * Sólo puede haber un comando mágico por celda
  * Un comando mágico debe ser lo primero en una celda

#### Lenguaje mágico
El lenguaje de comandos mágicos permite la ejecución de código en lenguajes distintos al predeterminado del notebook. En este curso, veremos los siguientes lenguajes mágicos:
* <strong><code>&#37;python</code></strong>
* <strong><code>&#37;sql</code></strong>

No es necesario añadir el lenguaje mágico para el tipo de notebook actualmente configurado.

Cuando cambiamos el lenguaje del notebook de Python a SQL más arriba, las celdas existentes escritas en Python tenían el valor del comando <strong><code>&#37;python</code></strong> añadido.

**NOTA**: En lugar de cambiar constantemente el lenguaje por defecto de un notebook, deberías quedarte con un lenguaje primario por defecto y sólo usar lenguaje mágico cuando sea necesario para ejecutar código en otro lenguaje.

```python
%python
print("Hello Python!")
```

#### **%run**
* Puedes ejecutar un notebook desde otro notebook usando el comando mágico **%run**
* Los notebooks a ejecutar se especifican con rutas relativas
* El notebook referenciado se ejecuta como si fuera parte del notebook actual, por lo que las vistas temporales y otras declaraciones locales estarán disponibles desde el notebook que realiza la llamada

Descomentar y ejecutar la siguiente celda generará el siguiente error:<br/>
**`Error in SQL statement: AnalysisException: Table or view not found: demo_tmp_vw`**

```python
%sql
-- SELECT * FROM demo_tmp_vw
```

Pero podemos declararla y un puñado de otras variables y funciones ejecutando esta celda:

```python
%run ../Includes/Classroom-Setup-01.2
```

El notebook **`../Includes/Classroom-Setup-01.2`** al que hemos hecho referencia incluye lógica para crear y **`USE`** un schema, así como para crear la temp view **`demo_temp_vw`**.

Podemos ver que esta temp view está ahora disponible en nuestra sesión de notebook actual con la siguiente consulta.

```python
%sql 
SELECT * FROM demo_tmp_vw
```

### **Databricks Utilities**
Los notebooks de Databricks incluyen un objeto **`dbutils`** que proporciona una serie de comandos de utilidad para configurar e interactuar con el entorno: <a href="https://docs.databricks.com/user-guide/dev-tools/dbutils.html" target="_blank">dbutils docs</a>

A lo largo de este curso, usaremos ocasionalmente **`dbutils.fs.ls()`** para listar directorios de archivos desde celdas Python.

```python
path = f"{DA.paths.datasets}"
dbutils.fs.ls(path)
```

### **display()**

Al ejecutar consultas SQL desde celdas, los resultados siempre se mostrarán en un formato tabular renderizado.

Cuando tenemos datos tabulares devueltos por una celda Python, podemos llamar a **`display`** para obtener el mismo tipo de vista previa.

Aquí, envolveremos el comando list anterior en nuestro sistema de archivos con **`display`**.

```python
path = f"{DA.paths.datasets}"
files = dbutils.fs.ls(path)
display(files)
```

El comando **`display()`** tiene las siguientes capacidades y limitaciones:
* Vista previa de los resultados limitada a 1000 registros
* Proporciona un botón para descargar los datos de los resultados como CSV
* Permite representar gráficos

### **Descargando Notebooks**

Existen varias opciones para descargar tanto notebooks individuales como colecciones de notesbooks.

Aquí verás el proceso para descargar este notebook y una colección de todos los notebooks de este curso.

#### **Descargar un Notebook**

Pasos:
* Haz clic en la opción **File** a la derecha de la selección del cluster en la parte superior del notebook.
* En el menú que aparece, sitúe el cursor sobre **Export** y, a continuación, seleccione **Source File**.

El notebook se descargará en su ordenador portátil. Se denominará con el nombre actual del notebook y tendrá la extensión de archivo correspondiente al lenguaje por defecto. Puedes abrir este notebook con cualquier editor de archivos y ver el contenido en bruto de los notebooks de Databricks.

Estos "source files" pueden ser cargados en cualquier workspace de Databricks.

#### **Descargar una colección de notebooks**

**NOTA**: Las siguientes instrucciones asumen que usted ha importado estos materiales usando **Repos**.

Pasos:
* Haga clic en el ![](https://files.training.databricks.com/images/repos-icon.png) **Repos** en la barra lateral izquierda.
  * Esto debería darte una vista previa de los directorios padre para este notebook
* En el lado izquierdo de la vista previa del directorio, en el centro de la pantalla, debería haber una flecha hacia la izquierda. Haz clic en ella para moverte hacia arriba en la jerarquía de archivos.
* Deberías ver un directorio llamado **Data Engineering with Databricks**. Haz clic en la flecha hacia abajo para abrir un menú.
* En el menú, sitúate sobre **Export** y selecciona **DBC Archive**.

El archivo DBC (Databricks Cloud) que se descarga contiene una colección comprimida de los directorios y notebooks de este curso. Los usuarios no deben intentar editar estos archivos DBC localmente, pero se pueden cargar de forma segura en cualquier workspace de Databricks para mover o compartir los contenidos de los notebooks.

**NOTA**: Al descargar una colección de DBCs, también se exportarán las vistas previas de los resultados y los gráficos. Al descargar source notebooks, sólo se guardará el código.
