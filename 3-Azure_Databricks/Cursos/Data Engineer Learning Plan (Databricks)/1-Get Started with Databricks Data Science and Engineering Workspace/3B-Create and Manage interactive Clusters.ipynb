### **Create and Manage Interactive Clusters**

Un clúster Databricks es un conjunto de recursos de computación y configuraciones en los que se ejecutan cargas de trabajo de ingeniería de datos, ciencia de datos y análisis de datos, como pipelines ETL de producción, análisis de streaming, análisis ad hoc y machine learning. Estos workloads se ejecutan como un conjunto de comandos en un notebook o como un job automatizado. 

Databricks distingue entre all-purpose clusters y job clusters. 
- Los all-purpose clusters se utilizan para analizar datos de forma colaborativa mediante notebooks interactivos.
- Los job clusters se utilizan para ejecutar jobs automatizados rápidos y robustos.

Esta demo cubrirá la creación y gestión de all-purpose Databricks clusters usando el Databricks Data Science & Engineering Workspace. 

### **Objetivos de aprendizaje**
Al final de esta lección, usted debe ser capaz de:
* Utilizar la UI de Clusters para configurar y desplegar un cluster.
* Editar, terminar, reiniciar y eliminar clusters

### **Crear un Cluster**

Dependiendo del workspace en el que esté trabajando actualmente, puede o no tener privilegios de creación de clusters. 

Las instrucciones de esta sección asumen que **tiene** privilegios de creación de cluster, y que necesita desplegar un nuevo cluster para ejecutar las lecciones de este curso.

**NOTA**: Consulte con su instructor o con un administrador de la plataforma para confirmar si debe o no crear un nuevo cluster o conectarse a un cluster que ya ha sido desplegado. Las políticas de clúster pueden afectar a sus opciones de configuración del clúster. 

Pasos:
1. Utilice la barra lateral izquierda para navegar a la página **Compute** haciendo clic en el icono ![compute](https://files.training.databricks.com/images/clusters-icon.png)
1. Haga clic en el botón azul **Create Cluster**.
1. Para el **Cluster name**, utiliza tu nombre para que puedas encontrarlo fácilmente y el instructor pueda identificarlo fácilmente si tienes problemas
1. Establece el **Cluster mode** a **Single Node** (este modo es necesario para realizar este curso)
1. Utiliza la **Databricks runtime version** recomendada para este curso
1. Deje marcadas las casillas para la configuración por defecto en las **Autopilot Options**.
1. Haz clic en el botón azul **Create Cluster**.

**NOTA:** Los clusters pueden tardar varios minutos en desplegarse. Una vez que haya terminado de desplegar un clúster, no dude en seguir explorando la interfaz de usuario de creación de clústeres.

### <img src="https://files.training.databricks.com/images/icon_warn_24.png"> Se requiere un clúster Single-Node para este curso.
**IMPORTANTE:** Este curso requiere que ejecutes notebooks en un cluster single-node. 

Siga las instrucciones anteriores para crear un clúster que tenga el **Cluster mode** configurado en **`Single Node`**.

### **Gestionar Clusters**

Una vez creado el clúster, vuelva a la página **Compute** para ver el clúster.

Seleccione un cluster para revisar la configuración actual. 

Haga clic en el botón **Edit**. Tenga en cuenta que la mayoría de los ajustes se pueden modificar (si dispone de permisos suficientes). Para cambiar la mayoría de los ajustes será necesario **reiniciar** los clústeres en ejecución.

**NOTA**: Utilizaremos nuestro cluster en la siguiente lección. Reiniciar, terminar o eliminar tu cluster puede retrasarte mientras esperas a que se desplieguen nuevos recursos.

### **Restart, Terminate, y Delete**

Ten en cuenta que aunque **Restart**, **Terminate**, y **Delete** tienen efectos diferentes, todos comienzan con un evento de terminación del cluster. (Los clústeres también terminarán automáticamente debido a la inactividad suponiendo que se utilice esta configuración).

Cuando un cluster termina, todos los recursos de la nube actualmente en uso son borrados. Esto significa:
* Se purgarán las máquinas virtuales asociadas y la memoria operativa.
* Se eliminará el volumen de almacenamiento adjunto
* Se eliminarán las conexiones de red entre nodos

En resumen, todos los recursos previamente asociados al entorno informático se eliminarán por completo. Esto significa que **cualquier resultado que necesite ser conservado debe ser guardado en una ubicación permanente**. Tenga en cuenta que no perderá su código, ni los archivos de datos que haya guardado adecuadamente.

El botón **Restart** nos permitirá reiniciar manualmente nuestro cluster. Esto puede ser útil si necesitamos limpiar completamente la caché del cluster o deseamos reiniciar completamente nuestro entorno de computación (Por ejemplo, cree una BD y una tabla, y al reiniciar esto se borró).

El botón **Terminate** nos permite detener nuestro cluster. Mantenemos la configuración de nuestro cluster, y podemos utilizar el botón **Restart** para desplegar un nuevo conjunto de recursos en la nube utilizando la misma configuración.

El botón **Delete** detendrá nuestro cluster y eliminará la configuración del cluster.
