##### DerivedColumn

- La función "instr" encuentra la posición (el primer caracter que encuentre) de la subcadena dentro de una cadena. 
  Se devuelve 0 si no se encuentra.

Ejemplos:
1. instr('dumbo', 'mbo') -> 3
2. instr('microsoft', 'o') -> 5
3. instr('bueno', 'malo') -> 0

- La función "substr" extrae de una posición una subcadena de una longitud determinada. Si se omite la longitud, se toma 
  por defecto el final de la cadena.

Ejemplos:
1. substring('Cat in the hat', 5, 2) -> 'in'
2. substring('Cat in the hat', 5, 100) -> 'in the hat'
3. substring('Cat in the hat', 5) -> 'in the hat'
4. substring('Cat in the hat', 100, 100) -> ''
